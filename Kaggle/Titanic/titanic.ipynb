```python
# 利用随机森林模型预测
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import SelectKBest
```

```python
train = pd.read_csv('Resources/my_train.csv')
test = pd.read_csv('Resources/my_test.csv')
origin_test = pd.read_csv('Resources/test.csv')
# 进行onehot编码需要将train test合并，否则可能因为test缺少某个枚举项导致shape不同
all_data = pd.concat([train, test])
all_data = pd.get_dummies(all_data)
train = all_data[all_data['Survived'].notnull()]
test_df = all_data[all_data['Survived'].isnull()].drop('Survived',axis=1)
X = train.as_matrix()[:, 1:]
y = train.as_matrix()[:, 0]
test = test_df.as_matrix()
print(X.shape)
print(test.shape)
```

```python
# 用网格搜索自动化选取参数
# 一开始运行这块代码sklearn抛出了Unknown label type: 'continuous'错误，这是由于test
pipe = Pipeline([('select', SelectKBest(k=20)),
                ('classify', RandomForestClassifier(random_state=10, max_features='sqrt'))])
param_test = {
    'classify__n_estimators': list(range(20, 50, 2)),
    'classify__max_depth': list(range(3, 60, 3))
}
gsearch = GridSearchCV(estimator=pipe, param_grid=param_test, scoring='roc_auc', cv=10)
gsearch.fit(X, y)
print(gsearch.best_params_, gsearch.best_score_)
```

```python
# 开始训练模型
from sklearn.pipeline import make_pipeline
select = SelectKBest(k=20)
clf = RandomForestClassifier(random_state=10, warm_start=True, n_estimators=26, max_depth=6, max_features='sqrt')
pipeline = make_pipeline(select, clf)
pipeline.fit(X, y)
```

```python
# 交叉验证
from sklearn import cross_validation, metrics
import numpy as np
cv_score = cross_validation.cross_val_score(pipeline, X, y, cv=10)
# 均值和标准差
print("cv score: Mean: %.7g | Std - % .7g" % (np.mean(cv_score), np.std(cv_score)))
```

```python
# 预测
print(test.shape)
print(X.shape)
pred = pipeline.predict(test)
submission = pd.DataFrame({'PassengerId': origin_test['PassengerId'], 'Survived': pred.astype(np.int32)})
submission.to_csv("Resources/my_submission.csv", index=False)
```
