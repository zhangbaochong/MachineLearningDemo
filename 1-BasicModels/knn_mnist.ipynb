```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# 获得mnist数据
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("../Resources/mnistdata", one_hot=True)
trainimg   = mnist.train.images
trainlabel = mnist.train.labels
# 测试集取200个
test_num = 200
testimg, testlabel = mnist.test.next_batch(test_num)
print(trainimg.shape, trainlabel.shape)
print(testimg.shape, testlabel.shape)
```

```python
# create graph
x_train = tf.placeholder(tf.float32, [None, 784])
x_test = tf.placeholder(tf.float32, [784])

# L1距离
distance = tf.reduce_sum(tf.abs(tf.add(x_train, tf.negative(x_test))), reduction_indices=1)

pred = tf.argmin(distance, 0)

acc = 0

init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(test_num):
        # Get nearest neighbor
        nn_index = sess.run(pred, feed_dict={x_train: trainimg, x_test: testimg[i, :]})
        print("Test", i, "Prediction:", np.argmax(trainlabel[nn_index]), "True Class:", np.argmax(testlabel[i]))
        # 计算精度
        if np.argmax(trainlabel[nn_index]) == np.argmax(testlabel[i]):
            acc += 1.0 / test_num
    print("Accuracy : %s" % acc)
```
